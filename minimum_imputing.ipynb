```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"https://www.piworks.net/Upload/Document/Original/country_vaccination_stats.csv")
df.head()
```

```python
#I am checking for any country with only NaN entries, only Kuwait is so
df.groupby(["country"]).sum()
```

```python
minimum_vax = df.groupby(["country"]).min().drop(["date", "vaccines"],axis="columns").T
minimum_vax
minimum_vax["Argentina"][0]
```

```python
countries = ['Argentina', 'Austria', 'Bahrain', 'Belgium', 'Brazil', 'Bulgaria',
       'Canada', 'Chile', 'China', 'Costa Rica', 'Croatia', 'Cyprus',
       'Czechia', 'Denmark', 'Ecuador', 'England', 'Estonia', 'Finland',
       'France', 'Germany', 'Gibraltar', 'Greece', 'Hungary', 'Iceland',
       'India', 'Indonesia', 'Ireland', 'Isle of Man', 'Israel', 'Italy',
       'Kuwait', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Mexico',
       'Netherlands', 'Northern Ireland', 'Norway', 'Oman', 'Panama', 'Poland',
       'Portugal', 'Romania', 'Russia', 'Saudi Arabia', 'Scotland', 'Serbia',
       'Seychelles', 'Singapore', 'Slovakia', 'Slovenia', 'Spain', 'Sweden',
       'Switzerland', 'Turkey', 'United Arab Emirates', 'United Kingdom',
       'United States', 'Wales']
new_df = []
for country_ in countries:
    q=df.loc[df["country"] == country_].fillna(minimum_vax[country_][0])
    new_df.append(q)
```

```python
df.loc[df["country"] == country_].fillna(minimum_vax[country_][0])
```

```python
new_df = pd.concat(new_df)
```

```python
new_df.info()
#Only remaning one is Kuwait because it had no median values to begin with as we know
```

```python
new_df["daily_vaccinations"] = new_df["daily_vaccinations"].fillna(0)
```

```python
new_df.info()
#we have 0 null values remaining
```

```python
top3 = new_df.groupby(["country"]).median().sort_values(["daily_vaccinations"],ascending=False).iloc[0:3]
top3
```
